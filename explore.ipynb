```python
KI67_CUTOFF = 20

FEATURES = [
    "sex",
    "onsetage",
    "menop",
    "patientclass",
    "histopat",
    "grade",
    "tnm_ct",
    "tnm_cn",
    "er",
    "pr",
    "her2ihc",
]

df = df_clinical.copy()

for column in ["ki67","onsetage","er","pr"]:
    if column in df.columns:
        df[column] = pd.to_numeric(df[column], errors="coerce")

df["target"] = pd.Series(pd.NA, index=df.index, dtype="Int64")
if "ki67" in df.columns:
    df.loc[df["ki67"].notna() & (df["ki67"] >= KI67_CUTOFF), "target"] = 1
    df.loc[df["ki67"].notna() & (df["ki67"] <  KI67_CUTOFF), "target"] = 0

selected_features = [column for column in FEATURES if column in df.columns]
cols = ["record_id"] + selected_features + ["target"]
df_ki = df[cols].copy()

df_ki = df_ki.dropna(subset=selected_features + ["target"])

df_ki = df_ki.drop(columns=[column for column in ["ki67","ki67_2"] if column in df_ki.columns], errors="ignore")

for column in selected_features:
    if pd.api.types.is_float_dtype(df_ki[column]) and df_ki[column].dropna().mod(1).eq(0).all():
        df_ki[column] = df_ki[column].astype("Int64")

print("df_ki :", df_ki.shape, "| positives =", int(df_ki["target"].sum()))
display(df_ki.head())
```
