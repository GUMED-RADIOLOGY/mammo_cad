```python
KI67_CUTOFF = 20

FEATURES = [
    "sex",
    "onsetage",
    "menop",
    "patientclass",
    "histopat",
    "grade",
    "tnm_ct",
    "tnm_cn",
    "er",
    "pr",
    "her2ihc",
]

df = df_clinical.copy()

for column in ["ki67","onsetage","er","pr"]:
    if column in df.columns:
        df[column] = pd.to_numeric(df[column], errors="coerce")

df["target"] = pd.Series(pd.NA, index=df.index, dtype="Int64")
if "ki67" in df.columns:
    df.loc[df["ki67"].notna() & (df["ki67"] >= KI67_CUTOFF), "target"] = 1
    df.loc[df["ki67"].notna() & (df["ki67"] <  KI67_CUTOFF), "target"] = 0

selected_features = [column for column in FEATURES if column in df.columns]
cols = ["record_id"] + selected_features + ["target"]
df_ki = df[cols].copy()

df_ki = df_ki.dropna(subset=selected_features + ["target"])

df_ki = df_ki.drop(columns=[column for column in ["ki67","ki67_2"] if column in df_ki.columns], errors="ignore")

for column in selected_features:
    if pd.api.types.is_float_dtype(df_ki[column]) and df_ki[column].dropna().mod(1).eq(0).all():
        df_ki[column] = df_ki[column].astype("Int64")

print("df_ki :", df_ki.shape, "| positives =", int(df_ki["target"].sum()))
display(df_ki.head())
```

```python
# Group stats & hypothesis tests for Ki-67 target (>=20% vs <20%)

# Split features by type (continuous, categorical, ordinal)
continuous = [c for c in ["onsetage","er","pr"] if c in df_ki.columns]
categorical = [c for c in ["sex","menop","patientclass","histopat","grade","tnm_ct","tnm_cn","her2ihc"] if c in df_ki.columns]
ordinal_for_trend = [c for c in ["grade","tnm_ct","tnm_cn","her2ihc"] if c in df_ki.columns]

print("Target distribution:\n", df_ki["target"].value_counts(dropna=False), "\n")

# Continuous features: group stats + t-test / Mann–Whitney U
cont_rows = []
for col in continuous:
    g0 = df_ki.loc[df_ki["target"]==0, col].dropna()
    g1 = df_ki.loc[df_ki["target"]==1, col].dropna()

    # Normality check (need n>=3) - D'Agostino-Pearson if n>=20, Shapiro-Wilk if 3<=n<5000, else skip
    def normal_p(x):
        if len(x) >= 20:
            return stats.normaltest(x, nan_policy="omit").pvalue
        elif 3 <= len(x) <= 5000:
            return stats.shapiro(x).pvalue
        else:
            return np.nan
    p_norm0, p_norm1 = normal_p(g0), normal_p(g1)
    normal = (p_norm0 is not np.nan and p_norm0 > 0.05) and (p_norm1 is not np.nan and p_norm1 > 0.05)

    # Variance equality check (need n>1 in both) - Levene's test 
    p_levene = stats.levene(g0, g1, center="median").pvalue if len(g0)>1 and len(g1)>1 else np.nan

    # Test selection and execution (need n>0 in both) - t-test if normal, else Mann–Whitney U
    if normal and not np.isnan(p_levene):
        # t-test: Student if equal var, else Welch
        t_res = stats.ttest_ind(g0, g1, equal_var=(p_levene>0.05))
        test = "t-test (Student)" if p_levene>0.05 else "t-test (Welch)"
        stat, pval = float(t_res.statistic), float(t_res.pvalue)
    else:
        # Mann–Whitney U test
        if len(g0)>0 and len(g1)>0:
            u = stats.mannwhitneyu(g0, g1, alternative="two-sided")
            test, stat, pval = "Mann–Whitney U", float(u.statistic), float(u.pvalue)
        else:
            test, stat, pval = "N/A", np.nan, np.nan

    cont_rows.append({
        "feature": col,
        "n0": len(g0), "mean0": g0.mean(), "sd0": g0.std(ddof=1), "median0": g0.median(), "IQR0": g0.quantile(0.75)-g0.quantile(0.25),
        "n1": len(g1), "mean1": g1.mean(), "sd1": g1.std(ddof=1), "median1": g1.median(), "IQR1": g1.quantile(0.75)-g1.quantile(0.25),
        "normal_p0": p_norm0, "normal_p1": p_norm1, "levene_p": p_levene,
        "test": test, "stat": stat, "p": pval
    })

cont_summary = pd.DataFrame(cont_rows)
print("Continuous featues:\n")
display(cont_summary)

# Categorical features: group counts + Chi-square / Fisher exact test + Cramér's V
cat_rows = []
for col in categorical:
    tbl = pd.crosstab(df_ki[col], df_ki["target"])
    if tbl.shape[1] != 2:
        tbl = tbl.reindex(columns=sorted(tbl.columns), fill_value=0)

    chi2, p_chi2, dof, exp = stats.chi2_contingency(tbl, correction=False)
    exp_min = exp.min()

    if tbl.shape == (2,2) and exp_min < 5:
        orr, p = stats.fisher_exact(tbl.values)
        effect = f"Odds ratio={orr:.3g}"
        test_name = "Fisher exact (2x2)"
        stat_val = np.nan
    else:
        n = tbl.values.sum()
        phi2 = chi2 / n if n>0 else np.nan
        r, k = tbl.shape
        phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1)) if n>1 else np.nan
        k_corr = k - ((k-1)**2)/(n-1) if n>1 else np.nan
        r_corr = r - ((r-1)**2)/(n-1) if n>1 else np.nan
        denom = np.sqrt((min(k_corr-1, r_corr-1)))
        cramer_v = (np.sqrt(phi2_corr) / denom) if (denom and denom>0) else np.nan
        test_name, p, stat_val, effect = "Chi-square", float(p_chi2), float(chi2), f"Cramér's V={cramer_v:.3g}"

    cat_rows.append({
        "feature": col,
        "levels": tbl.shape[0],
        "min_expected": float(exp_min),
        "test": test_name,
        "stat": stat_val,
        "p": float(p),
        "effect": effect
    })

cat_summary = pd.DataFrame(cat_rows)
display(cat_summary.sort_values("p"))

# Ordinal features: trend test (Spearman correlation)
trend_rows = []
for col in ordinal_for_trend:
    x = pd.to_numeric(df_ki[col], errors="coerce")
    y = df_ki["target"]
    mask = x.notna() & y.notna()
    if mask.sum() >= 3:
        rho, p = stats.spearmanr(x[mask], y[mask])
        trend_rows.append({"feature": col, "spearman_rho": float(rho), "p": float(p), "n": int(mask.sum())})

if trend_rows:
    trend_summary = pd.DataFrame(trend_rows).sort_values("p")
    display(trend_summary)
```
